```python
import cv2
import numpy as np
from deoldify import device
from deoldify.device_id import DeviceId
from deoldify.visualize import *
import torch
import os
from pathlib import Path
import warnings
warnings.filterwarnings("ignore")
```

```python
def colorize_image(input_image, output_image, render_factor=35):
    input_image = str(Path(input_image).absolute())
    output_image = str(Path(output_image).absolute())
    if not os.path.exists(input_image):
        raise FileNotFoundError("Giris goruntusu bulunamadi")
    device.set(device=DeviceId.GPU0 if torch.cuda.is_available() else DeviceId.CPU) 
    print("Model Yukleniyor...")
    result = get_image_colorizer(artistic=True)
    print("Renklendirme islemi basliyor")
    result.plot_transformed_image(path=input_image, render_factor=35, compare=True)
    print("Renklendirme islemi Bitti")
    

my_input_img = "stalin.jpg"
my_output_img = "colorized_output_img.png"
colorize_image(my_input_img, my_output_img, render_factor=35)
```
